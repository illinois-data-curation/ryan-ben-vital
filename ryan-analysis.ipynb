```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("integrated.csv")
df
```

```python
df.corr()
```

```python
sns.heatmap(df.corr(), vmin=-1, vmax=1, cmap='RdBu')
```

```python
fig, ax1 = plt.subplots()

ax1.plot(df['year'], df['majinjury'], label="Major Injury", color='red')
ax1.set_xlabel("Year")
ax1.set_ylabel("Fatalities")
ax1.set_title("Precipitation and Fatalities 2014-2024")

ax2 = ax1.twinx()
ax2.plot(df['year'], df['precipitation'], label="Precipitation", color='blue')
ax2.set_ylabel("Precipitation")
fig.legend()
```
